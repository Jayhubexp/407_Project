# Applying Pseudo Techniques for Visualization

### Objectives
- Convert grayscale images to pseudocolor using color map tchniques for better
visualization


```python
# Uploading Image folder
from google.colab import files
uploaded = files.upload()
```

```python
import cv2
import matplotlib.pyplot as plt
```

```python
# Step 2: Load the grayscale image
image_path = list(uploaded.keys())[0]  # Get the uploaded file's name
grayscale_image = cv2.imread(image_path, cv2.IMREAD_GRAYSCALE)

if grayscale_image is None:
    print("Error: Could not load the image. Please check the file path.")
else:
    print("Grayscale image loaded successfully.")
```

```python
colormaps = {
    "Jet": cv2.COLORMAP_JET,
    "Hot": cv2.COLORMAP_HOT,
    "Ocean": cv2.COLORMAP_OCEAN,
    "Rainbow": cv2.COLORMAP_RAINBOW,
    "Cool": cv2.COLORMAP_COOL
}

pseudocolored_images = {}
for cmap_name, cmap_code in colormaps.items():
    pseudocolored_images[cmap_name] = cv2.applyColorMap(grayscale_image, cmap_code)
```

```python
plt.figure(figsize=(15, 10))

# Display the original grayscale image
plt.subplot(2, 3, 1)
plt.imshow(grayscale_image, cmap='gray')
plt.title("Original Grayscale")
plt.axis("off")

# Display pseudocolored images
for i, (cmap_name, colored_image) in enumerate(pseudocolored_images.items(), start=2):
    plt.subplot(2, 3, i)
    rgb_image = cv2.cvtColor(colored_image, cv2.COLOR_BGR2RGB)
    plt.imshow(rgb_image)
    plt.title(f"Pseudocolor: {cmap_name}")
    plt.axis("off")

plt.tight_layout()
plt.show()
```
